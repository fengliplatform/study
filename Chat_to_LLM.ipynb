**1 Chat with downloaded facebook/opt-125m Model via vLLM**

```python
!pip install vllm
```

```python
from vllm import LLM, SamplingParams
```

```python
# download model from HF using HF_TOKEN from notebook secret
llm = LLM(model="facebook/opt-125m")
```

```python
prompts = [
    "Hello, my name is",
    "The president of the United States is",
    "The capital of France is",
    "The future of AI is",
]
sampling_params = SamplingParams(temperature=0.8, top_p=0.95)

outputs = llm.generate(prompts, sampling_params)
# Print the outputs.
for output in outputs:
    prompt = output.prompt
    generated_text = output.outputs[0].text
    print(f"Prompt: {prompt!r}, Generated text: {generated_text!r}")
```

**2 Chat with my OAI hosted at OAI site**

```python
!pip install openai
```

```python
from openai import OpenAI

# get colab notebook secret for OAI sevice user api key in default porject
from google.colab import userdata
oai_api_key = userdata.get('OPENAI_API_KEY')


client = OpenAI(
    api_key = oai_api_key
)
```

```python
# Chat with my OAI
completion = client.chat.completions.create(
    model="gpt-4o-mini",
    #store=True,
    messages=[
        {"role": "user", "content": "how to make banana bread?"}
    ]
)
print(completion.choices[0].message)
```

```python
from langchain.llms import OpenAI
from google.colab import userdata
oai_api_key = userdata.get('OPENAI_API_KEY')


client = OpenAI(
    api_key = oai_api_key
)
client.invoke("hello how are you?")
```

```python
from langchain.chat_models import ChatOpenAI

model = ChatOpenAI(api_key = oai_api_key, model="gpt-4o-mini")

model.invoke("Hello, world!")

```
